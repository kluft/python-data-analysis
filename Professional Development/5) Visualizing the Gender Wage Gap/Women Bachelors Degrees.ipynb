# Percentage of Women with Bachelors Degrees in the US

In this analysis we are going to look at the percentage of women with Bachelors degrees in the US. Furthermore, we will output this data visually so we can easily analyze the data in one figure for 17 different Bachelors degrees.





```python
# Import the libraries we need in order to make the plots 

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Read in the file we want
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

# Set the line colour
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# These are the degrees we want to look at visually to compare gender gaps
# These are separated by different categories - STEM, Liberal Arts, and others.
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English','Communications and Journalism', 'Art and Performance','Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business','Architecture']

# Combine all lists into a list of lists
all_cats = [stem_cats, lib_arts_cats, other_cats]

# Set the size of the figure
fig = plt.figure(figsize=(12, 30))

# Set which plot is the starting point for each column
beg_num = 1

# The for loop searches through all lists in the all_cats list of lists.
for cat in all_cats:
    # Resets the list_val variable to 0 for the next list that needs to be run through
    # This variable searches through each of the Bachelors category lists to identify the specific major
    list_val = 0
    
    for sp in range(beg_num,19,3):
        # This try block is necessary for when we get to the last element of the second list. The second list has
        # only 5 elements instead of 6 so this stops the code from throwing an error.
        try:
            # All of this code specifies the exact details we want for the graphs as they are created
            
            # Specifies which subplot is currently being worked on
            ax = fig.add_subplot(6,3,sp)
            
            # Applies the data to the X and Y axes
            ax.plot(women_degrees['Year'], women_degrees[cat[list_val]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[cat[list_val]], c=cb_orange, label='Men', linewidth=3)
            
            # Hides the spines on the graphs
            ax.spines["right"].set_visible(False)    
            ax.spines["left"].set_visible(False)
            ax.spines["top"].set_visible(False)    
            ax.spines["bottom"].set_visible(False)
            
            # Sets the x limits between 1968 - 2011
            ax.set_xlim(1968, 2011)
            
            # Sets the y limits between 0 - 100
            ax.set_ylim(0,100)
            
            # Sets the tile of the graph
            ax.set_title(cat[list_val])
            
            # Turns off all tick parameters except the left side.
            ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
            
            # Sets the tick marks to only 0 and 100 (the top and bottom of the line)
            ax.set_yticks([0,100])
            
            # Creates a horizontal line in the middle
            ax.axhline(50,c=(171/255,171/255,171/255),alpha=.3)
            
            # These if statements 
            if sp == 1:
                ax.text(2002, 87, 'Men')
                ax.text(2002, 8, 'Women')
            elif sp == 16:
                ax.text(2002, 62, 'Men')
                ax.text(2001, 35, 'Women')
                ax.tick_params(labelbottom='on')
            elif sp == 2:
                ax.text(2001, 32, 'Men')
                ax.text(2005, 64, 'Women')
            elif sp == 3:
                ax.text(2001, 20, 'Men')
                ax.text(2005, 80, 'Women')
            elif sp == 14:
                ax.tick_params(labelbottom='on')   
            elif sp == 18:
                ax.text(2001, 63, 'Men')
                ax.text(2005, 35, 'Women')
                ax.tick_params(labelbottom='on')
            list_val += 1
        except:

            # Hide all graph data from the graph that threw an error
            ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off',labelleft='off')
            ax.spines["right"].set_visible(False)    
            ax.spines["left"].set_visible(False)
            ax.spines["top"].set_visible(False)    
            ax.spines["bottom"].set_visible(False)           
    
    # Increase the beginning start number for the top part of the graphs
    beg_num += 1
    
# Make sure we save a file in the containing folder this code is run
plt.savefig('gender_degrees.png')

# Show the plot
plt.show()
```

Using this visual method we can easily scan each of these graphs to get a broad overview of what information we are dealing with and which majors we should take a closer look at. These graphs can be incredibly useful and powerful when having to sift through a lot of data.
