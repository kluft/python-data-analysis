```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]

yes_no = {'Yes':True,'No':False}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
import numpy as np
seen_1_map = ['Star Wars: Episode I  The Phantom Menace','Star Wars: Episode II  Attack of the Clones','Star Wars: Episode III  Revenge of the Sith','Star Wars: Episode IV  A New Hope','Star Wars: Episode V The Empire Strikes Back','Star Wars: Episode VI Return of the Jedi']
star_wars[star_wars.columns[3:9]] = star_wars[star_wars.columns[3:9]].isin(seen_1_map)

star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1","Unnamed: 4":"seen_2","Unnamed: 5":"seen_3","Unnamed: 6":"seen_4","Unnamed: 7":"seen_5","Unnamed: 8": "seen_6"
})
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1","Unnamed: 10":"ranking_2","Unnamed: 11":"ranking_3","Unnamed: 12":"ranking_4","Unnamed: 13":"ranking_5","Unnamed: 14": "ranking _6"
})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fav_movie_mean = star_wars[star_wars.columns[9:15]].mean()
print(fav_movie_mean)
labels = ["Star Wars 1","Star Wars 2","Star Wars 3","Star Wars 4","Star Wars 5","Star Wars 6"]
ax = fav_movie_mean.plot.bar()
ax.set_title("Favourite Star Wars Movie")
ax.set_xlabel("Star Wars Movie")
ax.set_ylabel("Mean value of Ranking")
ax.set_xticklabels(labels, rotation=35)
plt.show()
```

In the above graph we can quite easily see that Star Wars 5: The Empire Strikes back is favourite movie. We have to remember that the lower number here is a good thing as they are ranked from favourite being 1 to least favourite being 6. Another thing to note is that Star Wars 3 was the least favourite Star Wars movie.

```python
seen_sum = star_wars[star_wars.columns[3:9]].sum()

labels = ["Star Wars 1","Star Wars 2","Star Wars 3","Star Wars 4","Star Wars 5","Star Wars 6"]
ax = seen_sum.plot.bar()
ax.set_title("Star Wars Movies Seen")
ax.set_xlabel("Star Wars Movie")
ax.set_ylabel("Number of Movies Seen")
ax.set_xticklabels(labels, rotation=35)
plt.show()
```

In the above graph we can see that the 5th star wars film has the most seen. Now, both graphs we have examined have the 5th movie as the higest in each category. 

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

male_fav_movie_mean = males[males.columns[9:15]].mean()

labels = ["Star Wars 1","Star Wars 2","Star Wars 3","Star Wars 4","Star Wars 5","Star Wars 6"]
ax = male_fav_movie_mean.plot.bar()
ax.set_title("Male Favourite Star Wars Movie")
ax.set_xlabel("Star Wars Movie")
ax.set_ylabel("Mean value of Ranking")
ax.set_xticklabels(labels, rotation=35)
plt.show()
```

```python
males_seen_sum = males[males.columns[3:9]].sum()
labels = ["Star Wars 1","Star Wars 2","Star Wars 3","Star Wars 4","Star Wars 5","Star Wars 6"]
ax = males_seen_sum.plot.bar()
ax.set_title("Male Star Wars Movies Seen")
ax.set_xlabel("Star Wars Movie")
ax.set_ylabel("Number of Movies Seen")
ax.set_xticklabels(labels, rotation=35)
plt.show()
```

```python
female_fav_movie_mean = females[females.columns[9:15]].mean()

labels = ["Star Wars 1","Star Wars 2","Star Wars 3","Star Wars 4","Star Wars 5","Star Wars 6"]
ax = female_fav_movie_mean.plot.bar()
ax.set_title("Female Favourite Star Wars Movie")
ax.set_xlabel("Star Wars Movie")
ax.set_ylabel("Mean value of Ranking")
ax.set_xticklabels(labels, rotation=35)
plt.show()
```

```python
females_seen_sum = females[females.columns[3:9]].sum()
labels = ["Star Wars 1","Star Wars 2","Star Wars 3","Star Wars 4","Star Wars 5","Star Wars 6"]
ax = females_seen_sum.plot.bar()
ax.set_title("Female Star Wars Movies Seen")
ax.set_xlabel("Star Wars Movie")
ax.set_ylabel("Number of Movies Seen")
ax.set_xticklabels(labels, rotation=35)
plt.show()
```

The biggest difference we can make when looking at these graphs is that females have not seen as many Star Wars movies than males.
